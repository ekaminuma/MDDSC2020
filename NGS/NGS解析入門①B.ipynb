# [D] SNPEffによるSNP注釈


```
#-------[D-1:SNPEeff解析ツールをColabにダウンロード]----------------------------------------------------------------------
#
#
%%bash

curl -v -L http://sourceforge.net/projects/snpeff/files/snpEff_latest_core.zip > snpEff_latest_core.zip
unzip snpEff_latest_core.zip
```

```
#-------[D-2: SNPEeffツールが動くか確認]----------------------------------------------------------------------
#
# SNPeffはjavaプログラム
#
# ※colabは、javaがインストール済
#
#-------------------------------------------------
%%bash

cd snpEff
java -jar snpEff.jar
```

```
#-------[D-3: SNPEeffのヒトゲノム注釈データベースをダウンロード]----------------------------------------------------------------------
#
#  1000genomes当時のヒトゲノムのversionは「hg19」。
#  ファイルサイズが大きいので、ダウンロード処理に数分かかります。
#
#-------------------------------------------------------------
%%bash


cd snpEff
java -jar snpEff.jar download -v hg19
#java -jar snpEff.jar download -v GRCh37.75

```

```
#-------[D-4: ダウンロードしたファイルの確認]----------------------------------------------------------------------
#
#


%%bash


cd snpEff
ls -lt
echo "--------------------"
ls -lt /content/snpEff/data

```

```
#-------[D-5: queryとなるSNP情報(vcfファイル)を取得]-----------------------
#
#  NA18507個体のchr21のvcfファイルをColabに取得
#

%%bash


cd ./snpEff
pwd

wget https://data.bits.vib.be/pub/trainingen/NGSDNA2013/ex7-files/snpeff_chr21-results/gwas-chr21_common_NA18507.vcf

mv /content/snpEff/gwas-chr21_common_NA18507.vcf  NA18507tmp.vcf
ls -l
```

```
#-------[D-6: vcfファイルの行数カウントと中身確認]-----------------------
#
# wc -l でファイルの行数確認
# head -50　で最初の50行を表示, chr21のSNP情報記載のvcfと判る
#
#  #CHROM	POS	ID	REF	ALT	QUAL	FILTER	INFO	FORMAT	GAIIx-chr21-BWA.mem
#  chr21	9467416	.	C	T	86	.	SF=0,1	GT:PL:GQ	0/1:116,0,126:99
#  chr21	9467417	.	A	C	111	.	SF=0,1,2	GT:PL:GQ	0/1:141,0,95:98
#  chr21	9471670	.	A	G	37.8	.	SF=0,1,2	GT:PL:GQ	1/1:69,6,0:10
#
#------------------------------------------------------------------------------
%%bash


wc -l /content/snpEff/NA18507tmp.vcf

echo "--------------------"

head -120 /content/snpEff/NA18507tmp.vcf

```

```
#-------[D-7: SnpEffによる「NA18507tmp.vcf」の変異機能注釈]-----------------------
#
# SnpEffの実行には、時間がかかります(数分)
#
# SnpEffはjavaプログラムなので、実行にはjavaが必要です。
# Google Colabではデフォルトでjavaが入っています。
#
#
#
#------------------------------------------------------------------------------

%%bash

cd snpEff

java -Xmx4g -jar snpEff.jar -v -stats NA18507_chr21.html hg19 ./NA18507tmp.vcf > ./NA18507tmp.ann.vcf
```

SnpEffは3つの出力ファイルを生成する。

- バリアントとその注釈についての要約統計を含むHTMLファイル。
- 注釈付きのVCF
- 遺伝子あたりの変異型の数をまとめたテキストファイル。



```
#-------[D-8: SnpEff実行ログ(htmlファイル)の表示]-----------------------
#
# snpeffの解析結果htmlを表示
# ここでは、pythonコマンドを使います。
#
# snpeff結果が、下記URLのデータと同じか、確認して下さい。
# https://data.bits.vib.be/pub/trainingen/NGSDNA2013/ex7-files/snpeff_chr21-results/annotated_chr21_common_NA18507.vcf
#
#---------------------------------------------------------------

import codecs
import IPython.display as d


f=codecs.open("./snpEff/NA18507_chr21.html", 'r')
d.display_html(f.read(), raw=True)
```

```
#-------[D-9: 「NA18507tmp.ann.vcf」からmissense変異のみ抽出]-----------------------
#
#
# grepで「missense_variant」の注釈が付いたSNPを表示する。
#
# grepの「-n」オプションは、行番号表示。
#
#
#------------------------------------------------------------------------------

%%bash


grep -n 'missense_variant' /content/snpEff/NA18507tmp.ann.vcf
```
